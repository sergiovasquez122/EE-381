```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unif_pdf(a,b,x):
    f = (1/abs(b-a))*np.ones(np.size(x))
    return f
```

```python
def uniform_rv(a, b, n):
    # Generate the values of the RV X
    x = np.random.uniform(a, b, n)

    # Create bins and histogram
    nbins = 30
    edgecolor = 'w'
    bins = [float(x) for x in np.linspace(a, b, nbins+1)]
    h1, bin_edges = np.histogram(x,bins,density=True)

    # Define points on the horizontal axis

    be1=bin_edges[0:np.size(bin_edges)-1]
    be2=bin_edges[1:np.size(bin_edges)]
    b1=(be1+be2)/2
    barwidth=b1[1]-b1[0] # Width of bars in the bargraph
    plt.close('all')

    # PLOT THE BAR GRAPH
    fig1 = plt.figure(1)
    plt.bar(b1,h1, width=barwidth, edgecolor=edgecolor)

    # PLOT THE UNIFORM PDF
    f = unif_pdf(a, b, b1)
    plt.plot(b1, f, 'r')
    plt.title('Uniform random variable Simulation')
    plt.xlabel('n = '+ str(n))
    plt.ylabel('Probability')
    plt.show()

    # CALCULATE THE MEAN AND STANDARD DEVIATION
    mu_x = np.mean(x)
    sig_x = np.std(x)

    print('mu_x = ', mu_x)
    print('sig_x = ', sig_x)
```

```python
a = 1.0
b = 4.0
n = 10000
uniform_rv(a, b, n)
```

```python
def exp_pdf(beta, t):
    f = np.exp((-1 / beta) * t) * (1 / beta)
    return f
```

```python
def exponential_rv(beta, n):
    t = np.random.exponential(beta, n)
    nbins = 30
    edgecolor = 'w'
    bins = [float(t) for t in np.linspace(0, 100, nbins + 1)]
    h1, bin_edges = np.histogram(t, bins, density=True)

    # Define points on the horizontal axis

    be1 = bin_edges[0:np.size(bin_edges) - 1]
    be2 = bin_edges[1:np.size(bin_edges)]
    b1 = (be1 + be2) / 2
    barwidth = b1[1] - b1[0]  # Width of bars in the bargraph
    plt.close('all')

    # PLOT THE BAR GRAPH
    fig1 = plt.figure(1)
    plt.bar(b1, h1, width=barwidth, edgecolor=edgecolor)

    # PLOT THE UNIFORM PDF
    f = exp_pdf(beta, b1)
    plt.plot(b1, f, 'r')
    plt.title('Exponential RV Simulation')
    plt.xlabel('n = ' + str(n))
    plt.ylabel('Probability')
    plt.show()

    # CALCULATE THE MEAN AND STANDARD DEVIATION
    mu_t = np.mean(t)
    sig_t = np.std(t)

    print('mu_t = ', mu_t)
    print('sig_t = ', sig_t)
```

```python
beta = 40
n = 10000
exponential_rv(beta, n)
```

```python
def normal_pdf(mu, sig, z):
    f = np.exp(-(z - mu) ** 2 / (2 * sig ** 2)) / (sig * np.sqrt(2 * np.pi))
    return f
```

```python
def normal_rv(mu, sigma, n):
    x = np.random.normal(mu, sigma, n)
    nbins = 30
    edgecolor = 'w'
    bins = [float(x) for x in np.linspace(1, 4, nbins+1)]
    h1, bin_edges = np.histogram(x,bins,density=True)

    # Define points on the horizontal axis
    be1=bin_edges[0:np.size(bin_edges)-1]
    be2=bin_edges[1:np.size(bin_edges)]
    b1=(be1+be2)/2
    barwidth = b1[1]-b1[0] # Width of bars in the bargraph
    plt.close('all')

    # PLOT THE BAR GRAPH
    fig1 = plt.figure(1)
    plt.bar(b1,h1, width=barwidth, edgecolor=edgecolor)

    # PLOT THE UNIFORM PDF
    f = normal_pdf(mu, sigma, b1)
    plt.plot(b1, f, 'r')
    plt.title('Normal RV Simulation')
    plt.xlabel('n = ' + str(n))
    plt.ylabel('Probability')
    plt.show()

    # CALCULATE THE MEAN AND STANDARD DEVIATION
    mu_x = np.mean(x)
    sig_x = np.std(x)

    print('mu_x = ', mu_x)
    print('sig_x = ', sig_x)
```

```python
mu = 2.5
sigma = 0.75
n = 10000
normal_rv(mu, sigma, n)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def generate_plot(a, b, nbooks, N):
    mu_x = (a + b) / 2
    sig_x = np.sqrt((b - a)**2/12)
    X = np.zeros((N,1))
    for k in range(0,N):
        x = np.random.uniform(a, b, nbooks)
        w = np.sum(x)
        X[k] = w
    nbins = 30
    edgecolor = 'w'
    bins = [float(x) for x in np.linspace(nbooks*a, nbooks*b, nbins + 1)]
    h1, bin_edges = np.histogram(X, bins, density = True)
    # Define points on the horizontal axis
    be1 = bin_edges[0:np.size(bin_edges) - 1]
    be2 = bin_edges[1:np.size(bin_edges)]
    b1 = (be1 + be2) / 2
    barwidth = b1[1] - b1[0] # Width of bars in the bargraph
    plt.close('all')

    fig1 = plt.figure(1)
    plt.bar(b1, h1, width = barwidth, edgecolor = edgecolor)

    def gaussian(mu, sig, z):
        f = np.exp(-(z - mu)**2 / (2*sig**2)) / (sig * np.sqrt(2*np.pi))
        return f

    f = gaussian(mu_x * nbooks, sig_x * np.sqrt(nbooks), b1)
    plt.plot(b1, f, 'r')
    plt.title('PDF of book stack height and comparison with Gaussian')
    plt.xlabel('Book stack height for n = ' + str(nbooks) + " books")
    plt.ylabel('Probability')
    plt.show()
```

```python
a = 1
b = 4
nbooks = 1
N = 10000
generate_plot(a, b, nbooks, N)
```

```python
nbooks = 5
generate_plot(a, b, nbooks, N)
```

```python
nbooks = 15
generate_plot(a, b, nbooks, N)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
beta = 40
carton_size = 24
N = 10000

X = np.zeros((N,1))
mu = beta * carton_size
sigma = beta * np.sqrt(carton_size)
for k in range(0,N):
    x = np.random.exponential(beta, carton_size)
    w = np.sum(x)
    X[k] = w
nbins = 30
edgecolor = 'w'
bins = [float(x) for x in np.linspace(0, 200, nbins + 1)]
h1, bin_edges = np.histogram(X, nbins, density = True)
# Define points on the horizontal axis
be1 = bin_edges[0:np.size(bin_edges) - 1]
be2 = bin_edges[1:np.size(bin_edges)]
b1 = (be1 + be2) / 2
barwidth = b1[1] - b1[0] # Width of bars in the bargraph
plt.close('all')

fig1 = plt.figure(1)
plt.bar(b1, h1, width = barwidth, edgecolor = edgecolor)

def gaussian(mu, sig, z):
    f = np.exp(-(z - mu)**2 / (2*sig**2)) / (sig * np.sqrt(2*np.pi))
    return f

f = gaussian(mu, sigma, b1)
plt.plot(b1, f, 'r')
plt.title('Random Variable of Exponential Distribution (PDF)')
plt.xlabel('Lifetime of a battery(days)')
plt.ylabel('Probability')
plt.show()
```

```python
plt.figure(2)
h1 = np.cumsum(h1 * barwidth)
CDF = np.cumsum(f * barwidth)
plt.bar(b1, h1, width = barwidth, edgecolor = edgecolor)
plt.plot(b1, CDF, 'r')
plt.title('Random Variable of Exponential Distribution (CDF)')
plt.xlabel('Total Lifetime (days)')
plt.ylabel('Probability')
plt.show()
```
